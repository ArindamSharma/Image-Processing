```
import cv2 
import numpy as np 
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
%matplotlib inline 
```

Loading the basic image we will be working with

```
def display_g(img):
  plt.imshow(img,cmap='gray')
def display(img):
  plt.imshow(img)
```

```
ori_img = cv2.imread("gate2.png")
img = cv2.cvtColor(ori_img,cv2.COLOR_BGR2RGB);
```

```
plt.imshow(img)
plt.show()
```

This is what the image looks like in gray scale
(We can see that most of the water fades away, lets apply some threshold methods and see the results)

```
gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.imshow(gray,cmap='gray')
```

The way thresh truncated works is that for any pixel value greater than thresh_hold, it is made to max_val (in our case 240)

```
ret1, thresh1 = cv2.threshold(gray,120,240,cv2.THRESH_TRUNC)
display_g(thresh1)
```

Trying some adaptive thresholding (Seems to give good results as well)

```
thresh2 = cv2.adaptiveThreshold(gray,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,121,4)
display_g(thresh2)
```

Trying to combine the two results 

```
thresh3 = cv2.adaptiveThreshold(thresh1,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,151,4)
display_g(thresh3)
```

This shows to yeild promising results
